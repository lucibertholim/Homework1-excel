```python
import pandas as pd
import requests
from pprint import pprint
import numpy as np


response = requests.get("https://www150.statcan.gc.ca/n1/dai-quo/ssi/homepage/ind-econ.json").json()

num_provinces = np.arange(0, len(response['results']['geo']))
province_names = {}

for index in num_provinces:
   province_names[index] = response['results']['geo'][index]['label']['en']
province_growth = {}

for index in num_provinces:
   province_growth[response['results']['indicators'][index]['geo_code']] = float(str(response['results']['indicators'][index]['growth_rate']['growth']['en']).replace("%", ''))
name_df = pd.Series(province_names)
province_df = pd.Series(province_growth)

combined_df = pd.concat([name_df, province_df], axis = 1)
combined_df
renamed_df = combined_df.rename(columns = {0:'Province', 1:'Rate'})
indexed_df = renamed_df.set_index('Province')
indexed_df.plot(kind='bar')
```
